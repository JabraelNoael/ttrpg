```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
```

```python
samples = (
    [0.2] * 20 +
    [0.6] * 80 +
    [0.9] * 30
)

kde = gaussian_kde(samples)

x = np.linspace(0, 1, 1000)
y = kde(x)

plt.plot(x, y)
plt.title("Custom Kernel Density Estimate")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```
